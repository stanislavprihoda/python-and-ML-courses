```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
#funtional method
```

```python
plt.plot(x,y)
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#OO method
```

```python
fig=plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xtitle('TITLE')
```

```python
fig=plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.65,0.2,0.2,0.2])
axes1.plot(x,y)
axes1.set_title('BIG PLOT')
axes2.plot(y,x)
axes2.set_title('SMALL PLOT')
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig=plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X squared')
ax.plot(x,x**3,label='X cubed')

ax.legend()
```

```python
fig=plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#FF8C00',linewidth=3,alpha=0.5,linestyle='--',marker='o',markersize=10,markerfacecolor='yellow')
```

```python
fig=plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',linewidth=3,linestyle='--')
ax.set_xlim([0,1])
ax.set_ylim([0,1])
```
