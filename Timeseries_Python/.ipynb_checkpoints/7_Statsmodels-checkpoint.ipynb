### EWMA

```python
airline = pd.read_csv('_sources/TSA_COURSE_NOTEBOOKS/Data/airline_passengers.csv',index_col='Month')
```

```python
airline.dropna(inplace=True)
```

```python
airline.index = pd.to_datetime(airline.index)
```

```python
airline.index
```

#### SMA recap

```python
airline['6-month-SMA'] = airline["Thousands of Passengers"].rolling(window=6).mean()
```

```python
airline['12-month-SMA'] = airline["Thousands of Passengers"].rolling(window=12).mean()
```

```python
airline.plot()
```

#### EWMA

```python
airline["EWMA-12"] = airline["Thousands of Passengers"].ewm(span=12).mean()
```

```python
airline[["Thousands of Passengers","EWMA-12"]].plot()
```

### Holt Winters Methods

```python
df = pd.read_csv('_sources/TSA_COURSE_NOTEBOOKS/Data/airline_passengers.csv',index_col='Month',parse_dates=True)
```

```python
df.head(3)
```

```python
df.dropna(inplace=True)
```

```python
df.index
```

```python
df.index.freq = 'MS'
```

```python
df.index
```

#### Single exponential smoothing

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
```

```python
span = 12
alpha = 2/(span+1)
```

```python
df['EWMA'] = df["Thousands of Passengers"].ewm(alpha=alpha,adjust=False).mean()
```

```python
df.plot()
```

```python
df.head(3)
```

#### Double exponential smoothing

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
df["DES_add_12"] = ExponentialSmoothing(df["Thousands of Passengers"],trend='add').fit().fittedvalues.shift(-1)
```

```python
df["DES_mul_12"] = ExponentialSmoothing(df["Thousands of Passengers"],trend='mul').fit().fittedvalues.shift(-1)
```

```python
df.plot()
```

```python
df.iloc[:24].plot(figsize=(12,5))
```

```python
df.iloc[:12].plot(figsize=(12,5))
```

```python
df["TES_mul_12"] = ExponentialSmoothing(df["Thousands of Passengers"],trend='mul',seasonal='mul',seasonal_periods=12).fit().fittedvalues
```

```python
df.iloc[:6].plot(figsize=(12,5))
```

### Excercises

```python
df = pd.read_csv('_sources/TSA_COURSE_NOTEBOOKS/Data/EnergyProduction.csv',index_col=0,parse_dates=True)
```

```python
df.head()
```

```python
df.index
```

```python
df.index.freq = 'MS'
```

```python
df.plot(figsize=(12,5))
```

```python
df["SMA"] = df["EnergyIndex"].rolling(window=12).mean()
```

```python
df.plot(figsize=(12,5))
```

```python
span = 12
alpha = 2/(span+1)
```

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
```

```python
df["SES"] = SimpleExpSmoothing(df["EnergyIndex"]).fit(smoothing_level=alpha, optimized=False).fittedvalues.shift(-1)
```

```python
df.head()
```

```python
df.plot(figsize=(12,5))
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
df["TSA_add_12"] = ExponentialSmoothing(df["EnergyIndex"],trend='mul',seasonal="mul",seasonal_periods=12).fit().fittedvalues.shift(-1)
```

```python
df.plot(figsize=(12,5))
```

```python
df.iloc[:24].plot(figsize=(12,5))
```
