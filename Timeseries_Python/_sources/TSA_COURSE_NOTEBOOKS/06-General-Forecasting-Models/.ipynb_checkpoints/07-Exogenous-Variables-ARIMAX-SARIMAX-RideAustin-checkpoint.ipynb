# SARIMAX

## Seasonal AutoRegressive Integrated Moving Average with eXogenous regressors
So far the models we've looked at consider past values of a dataset and past errors to determine future trends, seasonality and forecasted values. We look now to models that encompass these non-seasonal (p,d,q) and seasonal (P,D,Q,m) factors, but introduce the idea that external factors (environmental, economic, etc.) can also influence a time series, and be used in forecasting.

## Perform standard imports and load datasets

```python
import pandas as pd
import numpy as np
%matplotlib inline

# Load specific forecasting tools
from statsmodels.tsa.statespace.sarimax import SARIMAX

from statsmodels.graphics.tsaplots import plot_acf,plot_pacf # for determining (p,q) orders
from statsmodels.tsa.seasonal import seasonal_decompose      # for ETS Plots
from pmdarima import auto_arima                              # for determining ARIMA orders

# Ignore harmless warnings
import warnings
warnings.filterwarnings("ignore")

# Load dataset
df = pd.read_csv('../Data/RideAustin.csv',index_col='Date',parse_dates=True)
df.index.freq = 'D'
```

### Inspect the data
In this section we're using the RideAustin dataset. RideAustin is a non-profit, municipal ridesharing service based in Austin, Texas, with a model comparing directly to that of Uber. The data covers rides requested from June 4, 2016 to February 7, 2017, just over 35 weeks. It shows a clear weekly cycle, with peak ride demand on Saturday and Sunday.
Data was made available by CEO Andy Tryba, at https://data.world/andytryba/rideaustin

```python
df.head()
```

### Plot the source data

```python
title='RideAustin Ridesharing Service'
ylabel='Rides per day'
xlabel='' # we don't really need a label here

ax = df['Rides'].plot(figsize=(12,5),title=title)
ax.autoscale(axis='x',tight=True)
ax.set(xlabel=xlabel, ylabel=ylabel);
```

The data provided shows a increasing trend from June to October, presumably as the business was gaining popularity. For our purposes we'll look at the span from Sunday, October 2 to Saturday, February 4 (18 weeks).

```python
df = df.loc['2016-10-02':'2017-02-04']
df.head()
```

## Look at precipitation data
We'll also look at daily precipitation, to see if it might correlate to ridership.

```python
title='Austin Daily Precipitation'
ylabel=''
xlabel=''

ax = df['PRCP'].plot(figsize=(12,5),title=title)
ax.autoscale(axis='x',tight=True)
ax.set(xlabel=xlabel, ylabel=ylabel);
```

### Run an ETS Decomposition

```python
result = seasonal_decompose(df['Rides'], model='add')
result.plot();
```

```python
from statsmodels.tsa.stattools import adfuller

def adf_test(series,title=None):
    """
    Pass in a time series and an optional title, returns an ADF report
    """
    print(f'Augmented Dickey-Fuller Test: {title}')
    result = adfuller(series.dropna(),autolag='AIC') # .dropna() handles differenced data
    
    labels = ['ADF test statistic','p-value','# lags used','# observations']
    out = pd.Series(result[0:4],index=labels)

    for key,val in result[4].items():
        out[f'critical value ({key})']=val
        
    print(out.to_string())          # .to_string() removes the line "dtype: float64"
    
    if result[1] <= 0.05:
        print("Strong evidence against the null hypothesis")
        print("Reject the null hypothesis")
        print("Data has no unit root and is stationary")
    else:
        print("Weak evidence against the null hypothesis")
        print("Fail to reject the null hypothesis")
        print("Data has a unit root and is non-stationary")
```

```python
adf_test(df['Rides'])
```

```python
df['Rides'].plot()
```

### Run <tt>pmdarima.auto_arima</tt> to obtain recommended orders
This may take awhile as there are a lot of combinations to evaluate.

```python
# For SARIMA Orders we set seasonal=True and pass in an m value
auto_arima(df['Rides'],seasonal=True,m=7).summary()
```

Excellent! This provides an ARIMA Order of (3,0,3) and a seasonal order of (0,0,2,7) Now let's train & test the SARIMA model, evaluate it, then compare the result to a model that uses an exogenous variable.
### Split the data into train/test sets

```python
len(df)
```

```python
# Set four weeks for testing
train = df.iloc[:98]
test = df.iloc[98:]
```

### Fit a SARIMA(3,0,3)(0,0,2,7) Model

```python
model = SARIMAX(train['Rides'],order=(3,0,3),seasonal_order=(0,0,2,7))
results = model.fit()
results.summary()
```

```python
# Obtain predicted values
start=len(train)
end=len(train)+len(test)-1
predictions = results.predict(start=start, end=end, dynamic=False).rename('SARIMA(3,0,3)(0,0,2,7) Predictions')
```

Passing <tt>dynamic=False</tt> means that forecasts at each point are generated using the full history up to that point (all lagged values).

For more information on these arguments visit https://www.statsmodels.org/stable/generated/statsmodels.tsa.arima_model.ARIMAResults.predict.html

```python
# Plot predictions against known values
title = 'RideAustin Ridesharing Service'
ylabel='Rides per day'
xlabel=''

ax = test['Rides'].plot(legend=True,figsize=(12,6),title=title)
predictions.plot(legend=True)
ax.autoscale(axis='x',tight=True)
ax.set(xlabel=xlabel, ylabel=ylabel);
```

### Evaluate the Model

```python
from sklearn.metrics import mean_squared_error

error = mean_squared_error(test['Rides'], predictions)
print(f'SARIMA(3,0,3)(0,0,2,7) MSE Error: {error:11.10}')
```

```python
from statsmodels.tools.eval_measures import rmse

error = rmse(test['Rides'], predictions)
print(f'SARIMA(3,0,3)(0,0,2,7) RMSE Error: {error:11.10}')
```

## Now add the exog variable

```python
model = SARIMAX(train['Rides'],exog=train['PRCP'],order=(3,0,3),seasonal_order=(0,0,2,7),enforce_invertibility=False)
results = model.fit()
results.summary()
```

```python
# Obtain predicted values
start=len(train)
end=len(train)+len(test)-1
exog_forecast = test[['PRCP']]  # requires two brackets to yield a shape of (28,1)
predictions = results.predict(start=start, end=end, exog=exog_forecast).rename('SARIMAX(3,0,3)(0,0,2,7) Predictions')
```

```python
# Plot predictions against known values
title = 'RideAustin Ridesharing Service'
ylabel='Rides per day'
xlabel=''

ax = test['Rides'].plot(legend=True,figsize=(12,6),title=title)
predictions.plot(legend=True)
ax.autoscale(axis='x',tight=True)
ax.set(xlabel=xlabel, ylabel=ylabel);
```

### Retrain the model on the full data, and forecast the future

```python
model = SARIMAX(df['Rides'],order=(0,1,3),seasonal_order=(1,0,1,12))
results = model.fit()
fcast = results.predict(len(df),len(df)+11,typ='levels').rename('SARIMA(0,1,3)(1,0,1,12) Forecast')
```

```python
# Plot predictions against known values
title = 'RideAustin Ridesharing Service'
ylabel='Rides per day'
xlabel=''

ax = df['Rides'].plot(legend=True,figsize=(12,6),title=title)
fcast.plot(legend=True)
ax.autoscale(axis='x',tight=True)
ax.set(xlabel=xlabel, ylabel=ylabel);
```
